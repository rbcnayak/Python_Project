# 04. Subsetting

```python
# Subsetting means creating a smaller data set from a whole data set using: 'df_snacks =  df_prods[df_prods['department_id']==19]' 
# What the parts mean: 'df_prods['department_id']==19" assigns value to be either True or False based on criteria
df_prods['department_id']==19
```

```python
# What the parts mean: 'df_prods[df_prods['department_id']==19]' only those values within df_prods that are true (this is the subset)
df_prods[df_prods['department_id']==19]
```

```python
# assigning the subset to a new dataframe, which we'll call 'df_snacks' 
df_snacks =  df_prods[df_prods['department_id']==19]
```

```python
df_snacks.head()
```

```python
# Subsetting, using location of column (loc)
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]
```

```python
df_snacks_2.head()
```

```python
# Subsetting, using .isin --handy when checking for multiple values, for instance if we were to look for more than 1 "department_id" 
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```

```python
df_snacks_3.head()
```
